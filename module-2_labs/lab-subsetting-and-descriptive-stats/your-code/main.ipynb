#### Find the minimum, mean and the maximum of all numeric columns for each Department.

##### Hint: Use functions from Data Manipulation lesson

```python
# your answer here
print("Mean:")
display(employee.groupby("Department").mean())
print("Max:")
display(employee.groupby("Department").max().select_dtypes(include=["float64","int"]))
print("Min:")
display(employee.groupby("Department").min().select_dtypes(include=["float64","int"]))
#employee.describe().loc[["min","max","50%"]]
```
