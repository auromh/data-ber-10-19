##  Imports

```python
import pandas as pd
import requests
import json
import matplotlib
%matplotlib inline
```

# Bitcoin market value for last 30 days, vs. USD, deliver a DF and a Plot

```python
base = 'https://api.coingecko.com/api/v3/'
```

```python
param_dict = { 'id':'bitcoin', 
              'vs_currency':'usd', 
              'days':30}
```

```python
url = base + 'coins/bitcoin/market_chart'
result = requests.get(url, params=param_dict)
r_bc_us=result.json()
```

```python
df_bc_us=pd.DataFrame(r_bc_us["prices"],columns=["time","price"])
df_bc_us
```

```python
df_bc_us["time"]=pd.to_datetime(df_bc_us["time"],unit="ms")
df_bc_us.set_index("time",inplace=True)
```

```python
df_bc_us.plot()
```
